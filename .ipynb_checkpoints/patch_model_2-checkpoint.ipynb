## Loss & Helper

```python
def get_stats(full_pred: torch.tensor, full_target: torch.tensor, path="Validation"):
    # [B, C, H, W]
    out = {}
    
    for i, label in enumerate(target_labels):
        pred = full_pred[:, 0, i].detach().cpu().numpy()
        target = full_target[:, 0, i].detach().cpu().numpy()
    
        TP = np.sum(np.logical_and(pred==1 , target==1), axis=(0,1,2))
        FP = np.sum(np.logical_and(pred==1 , target==0), axis=(0,1,2))
        FN = np.sum(np.logical_and(pred==0 , target==1), axis=(0,1,2))
        TN = np.sum(np.logical_and(pred==0 , target==0), axis=(0,1,2))
    
        out["{}/accuracy/{}".format(path, label)] = (TP + TN) / (TP + FP + FN + TN)
        out["{}/dice/{}".format(path, label)] = 2*TP / (2*TP + FP + FN)
        out["{}/sensitivity/{}".format(path, label)] = TP / (TP + FN)
        out["{}/specificity/{}".format(path, label)] = TN / (TN + FP)
    
    return out
```

```python
def save_metrics(tracker, metrics, epoch):
    out = {}
    for key in metrics[0].keys():
        l = [item[key] for item in metrics]
        tracker.add_scalar(key, np.mean(l, axis=0), epoch) # we mean the metrics
    return out
```

```python
def load_checkpoint(model, path, strict=True):
    checkpoint = torch.load(path, map_location=device)
    #tracker.set_epoch(checkpoint['epoch'])
    model.load_state_dict(checkpoint['model_state_dict'], strict=strict)
    model.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    print("Loaded checkpoint from {}".format(path))
```

```python
def save_checkpoint(model, path):
    torch.save({
        #'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': model.optimizer.state_dict(),
    }, path)
    print("Saved checkpoint to {}".format(path))
```

# Training

```python
def train_loop(model, loader, tracker, epoch):
    model.train()
    
    # trackings
    metrics = []
    
    # loop over the train load
    for batch, (x, y, m) in enumerate(loader):
        
        # setting gradient to zero
        model.optimizer.zero_grad()
        
        # load to model
        x = x.to(device)
        y = y.to(device)
        m = m.to(device)
        
        # prediction
        prediction = model(x)
        loss = (model.loss_function(prediction, y) * m).sum()
        
        # backpropagation
        loss.backward()
        model.optimizer.step()
        
        # track
        print("loss = {:.2f}".format(loss.item()))
        metrics.append({"loss/train": loss.item()})
    
    # save metrics
    save_metrics(tracker, metrics, epoch)
```

```python
def valid_loop(model, loader, tracker, epoch):
    model.eval()
    
    # trackings
    metrics = []
    
    # loop over the train load
    with torch.no_grad():
        for batch, (x, y, m) in enumerate(loader):
            
            # load to model
            x = x.to(device)
            y = y.to(device)
            m = m.to(device)
            
            # prediction
            prediction = model(x)
            loss = (model.loss_function(prediction, y) * m).sum()
            bp_prediction = (prediction >= 0)
            
            # track
            print("loss = {:.2f}".format(loss.item()))
            stats = get_stats(bp_prediction, y)
            stats["loss/valid"] = loss.item()
            metrics.append(stats)

    # save metrics
    save_metrics(tracker, metrics, epoch)
```

```python
def run(model, train_loader, valid_loader, epochs=100):
    
    # starts
    tracker = SummaryWriter()
    
    for epoch in range(epochs):
        
        print("TRAINING")
        train_loop(model, train_loader, tracker, epoch)
        
        if (epoch % VALID_EVERY) == 0:
            print("VALIDATION")
            valid_loop(model, valid_loader, tracker, epoch)
```

## Ensembling

### Pre-masking

```python
def get_filter(slices, filter_shape):
    w, h = filter_shape // 4
    n = np.zeros(filter_shape)
    n[w:-w, h:-h] = 1
    return np.array([scipy.ndimage.gaussian_filter(n, sigma=10) for s in range(slices)])
```

```python
filters = get_filter(len(target_labels), SHAPE[1:,])
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].imshow(filters[0]), axs[1].plot(filters[0, 32])
```

```python
def grid_indices(image, stride=(1, 46, 46), filters=(5, 96, 96)):
    
    # get the bounds
    minb = (0, 0, 0)
    maxb = np.array(image.shape) - np.array(filters)
    
    # create grid
    x, y, z = np.meshgrid(*zip([np.arange(minb[i], maxb[i], stride[i]) for i in range(3)]))
    
    return np.array([x.flatten(), y.flatten(), z.flatten()]).T + np.array(filters)//2
```

```python
def point_inference(model, img, mask, test_indices, target_shape, filter_shape, filters=None):
    
    # dataset
    test_set = BrainDataset(test_indices, img, mask, test=True, shape=SHAPE)
    test_loader = DataLoader(test_set, batch_size=BATCH_SIZE_TEST, 
                             num_workers=NUM_WORKERS, pin_memory=True, drop_last=False)

    # cut around
    sx, sy, sz = np.array(filter_shape)//2
    sum_img = np.zeros(target_shape)
    count_img = np.zeros(target_shape)
    
    # add ones filter
    if filters is None:
        filters = np.ones((target_shape[0], filter_shape[1], filter_shape[2]))
    
    for idx, (inp, m, coor) in enumerate(test_loader):
        
        # pass through model
        try:
            
            # load data
            inp = inp.to(device)
            m = m.to(device)
            
            # make prediction
            pred = model(inp) * m
            pred = pred.detach().cpu().numpy()
            xs, ys, zs = coor
            
            # loop over subjects
            for i in range(pred.shape[0]):
                
                # position
                val = sigmoid(pred[i, 0])
                x, y, z = xs[i], ys[i], zs[i]
                
                # sum
                sum_img[:, x, y-sy:y+sy, z-sz:z+sz] += (val * filters)
                count_img[:, x, y-sy:y+sy, z-sz:z+sz] += filters

            if idx % 100 == 0:
                print(idx)
        
        except Exception as e:
            print(e)

    return np.divide(sum_img, count_img, where=(count_img != 0))
```

```python
def get_test_stats(pred: torch.tensor, target: torch.tensor, label=""):
    # [B, C, H, W]
    TP = np.sum(np.logical_and(pred==1 , target==1), axis=(0,1,2))
    FP = np.sum(np.logical_and(pred==1 , target==0), axis=(0,1,2))
    FN = np.sum(np.logical_and(pred==0 , target==1), axis=(0,1,2))
    TN = np.sum(np.logical_and(pred==0 , target==0), axis=(0,1,2))
    
    return {
        "accuracy/{}".format(label): (TP + TN) / (TP + FP + FN + TN),
        "dice/{}".format(label): 2*TP / (2*TP + FP + FN),
        "sensitivity/{}".format(label): TP / (TP + FN),
        "specificity/{}".format(label): TN / (TN + FP)
    }
```
